# <u>Instant Noodles King: Exploratory data analysis of ramen reviews</u>

Everyone loves instant noodles for its convenience, low cost and taste. But what makes a instant noodles great? Are there particular factors that we can identify that make one brand of instant noodles conclusively better than another?

## 1 Executive Summary

On first glance, the country of origin plays a big role on the rating of instant ramen. However, further analysis shows that country alone is inadequate in predicting whether instant ramen is good or not. Even with multiple linear regression (MLR) using additional binary indicators like whether there are artificial flavors, whether it is spicy, vegetarian, fried or large, the model only has a predictive power of 0.15. KNeighbors Regression does not fare any better than MLR - in fact it performs worse (score of 0.03). 

Even though there is no functional predictive model that I arrived at, there are some insights about what noodles we should buy/ eat. Spicy, pork based ramen are, on average, likely to be tastier than other flavored counterparts. 

## 2 Problem Statement & Dataset Selected

<h3>2.1 Problem Statement</h3>
<br>
With globalization and ease of shipping across borders, consumer choice has exponentially increased. Given the sheer number of instant ramen options available, how do we know what makes an instant ramen better than another instant ramen?

<h3>2.2 Data Selected</h3>
<br>
<b>2.2.1 Context</b><br>
The Ramen Rater is a product review website for the hardcore ramen enthusiast (or "ramenphile"), with over 2500 reviews to date. This dataset is an export of "The Big List" (of reviews), converted to a CSV format.

<b>2.2.2 Content</b>
<br>
Each record in the dataset is a single ramen product review. Review numbers are contiguous: more recently reviewed ramen varieties have higher numbers. Columns in this dataset include Brand, Variety (the product name), Country, and Style (whether it is packaged in a cup, bowl or tray). Stars indicate the ramen quality, as assessed by the reviewer, on a 5-point scale


<u>Data Source:</u> https://www.kaggle.com/residentmario/ramen-ratings#ramen-ratings.csv

## 3 Methodology, Insights & Evaluation

<h3>3.1 Methodology</h3>
<br>
0) <u><i>Data cleaning</i></u> - upon inspection, a lot of the data types are objects instead of string/float/int so some conversion is necessary; also, further processing is needed to convert strings into categorical/binary variables such that they can be used in MLR; NaN values also needed to be managed
<br><br>
1) First attempt to <u><i>explore the data on a macro level</i></u> and attempt to arrive at broad generalizations about the dataset
<br><br>
2) If time permits, <u><i>delve deeper into particular countries</i></u> to attempt to explain disparities in ratings

<h3>3.2 Insights</h3>
<br>
- Based on reviews, on average <u><i>spicy ramen are more tasty</u></i> (3.76) than non-spicy ramen (3.61)
<br><br>
- On average, you are <u><i>probably best off with eating pork flavored ramen</u></i> as opposed to other flavors. The breakdown of the star ratings are as follows:<br>
<br>
-- <b>Chicken</b>: 3.42<br>
-- <b>Beef</b>: 3.51<br>
-- <b>Pork</b>: 3.70<br>
-- <b>Vegetarian</b>: 3.09<br>
-- <b>Artificial</b>: 3.27<br>
<br>
- Contrary to popular belief, Japan does not have an exclusive claim over the best instant ramen. The average rating for instant ramen from Indonesia (4.07), Malaysia (4.12) and Singapore (4.12) are in fact higher than the average for Japan (3.98). Japan, however, is indeed most represented (14%) followed by surprisingly USA (13%)
<br><br>
- The distribution plot of stars are bimodal instead of following a normal distribution. The vast majority of cup noodles are within the 3.5 or 5 range - cumulative probability is about 0.6. This means that when one randomly chooses an instant ramen, chances are one will get something at least above average
<br><br>
- Contrary to initial expectations, the length of the brand name or explanation of the variety of instant noodles plays little to not part in affecting the stars that the ramen receives from the consumer

<h3>3.3 Evaluation</h3>
<br>
- The generic model to predict stars of instant ramen unfortunately lacks predictive power with a score of 0.14 - 0.15
<br><br>
-- Even for the country specific models for USA, Singapore and Japan, they were all unfortunately lacking in predictive power. The score is even lower than the general model. Perhaps other indicators should be considered
<br><br>
- I did not manage to include all the brands as categorical variables or do a deeper country-level analysis but suspect that that will yield more meaningful results
<br><br>
- There are clear implications of brand on star rating of the ramen. Possible further analysis includes decomposing each brand based on its history (e.g. number of years) among other quantifiable indicators
<br><br>

```python
#I want to know if the stars follow a normal distribution
plt.figure(figsize=(16,10))
ramen_df4=ramen_df4[~ramen_df4.Stars.str.contains("Unrated")]
ramen_df4['Stars'] = ramen_df4['Stars'].astype(float)
sns.distplot(ramen_df4['Stars'])
#okay guess not
```

```python
#in addition to creating binary variables of country and type, I suspect length of brand name and variety might have an effect on rating
#the intuition is that the longer the name, the harder it is to remember

ramen_df4['brand_length']  = ramen_df4['Brand'].str.len()
ramen_df4['style_length']  = ramen_df4['Variety'].str.len()
ramen_df4['spicy']= ramen_df4['Variety'].copy()
ramen_df4['spicy'] = ramen_df4['spicy'].astype(str)

def spicy_replace(x):
    if ('Spicy' in x or 'hot' in x or 'Kimchi' in x or 'Hot' in x or 'Tom Yum' in x or 'Curry' in x or 'Chili' in x or 'Tantanmen' in x or 'Pedas' in x or 'Laksa' in x):
        return 1
    else:
        return 0
ramen_df4['spicy'] = ramen_df4['spicy'].apply(spicy_replace)

ramen_df4['premium']= ramen_df4['Variety'].copy()
def premium_replace(x):
    if ('Premium' in x or 'Special' in x or 'Signature' in x or 'Supreme' in x):
        return 1
    else:
        return 0
ramen_df4['premium'] = ramen_df4['premium'].apply(premium_replace)

ramen_df4['seafood']= ramen_df4['Variety'].copy()
def seafood_replace(x):
    if ('Seafood' in x or 'Shrimp' in x or 'Prawn' in x or 'Crab' in x or 'Palabok' in x):
        return 1
    else:
        return 0
ramen_df4['seafood'] = ramen_df4['seafood'].apply(seafood_replace)

ramen_df4['miso']= ramen_df4['Variety'].copy()
def miso_replace(x):
    if ('Miso' in x):
        return 1
    else:
        return 0
ramen_df4['miso'] = ramen_df4['miso'].apply(miso_replace)

#fried
ramen_df4['fried']= ramen_df4['Variety'].copy()
def fried_replace(x):
    if ('Goreng' in x or 'Yakisoba' in x or 'Stir Fry' in x or 'Stir Fried' in x or 'Dry' in x or 'Chow Mein' in x):
        return 1
    else:
        return 0
ramen_df4['fried'] = ramen_df4['fried'].apply(fried_replace)

#artificial
ramen_df4['artificial']= ramen_df4['Variety'].copy()
def artificial_replace(x):
    if ('Artificial' in x):
        return 1
    else:
        return 0
ramen_df4['artificial'] = ramen_df4['artificial'].apply(artificial_replace)

#vegetarian
ramen_df4['vegetarian']= ramen_df4['Variety'].copy()
def veg_replace(x):
    if ('Vegetarian' in x or 'Imitation' in x):
        return 1
    else:
        return 0
ramen_df4['vegetarian'] = ramen_df4['vegetarian'].apply(veg_replace)

#pork
ramen_df4['pork']= ramen_df4['Variety'].copy()
def pork_replace(x):
    if ('Pork' in x or 'Buta' in x or 'Tonkotsu' in x):
        return 1
    else:
        return 0
ramen_df4['pork'] = ramen_df4['pork'].apply(pork_replace)

#chicken
ramen_df4['chicken']= ramen_df4['Variety'].copy()
def chicken_replace(x):
    if ('Chicken' in x or 'Ayam' in x):
        return 1
    else:
        return 0
ramen_df4['chicken'] = ramen_df4['chicken'].apply(chicken_replace)

#beef
ramen_df4['beef']= ramen_df4['Variety'].copy()
def beef_replace(x):
    if ('Beef' in x or 'Gyu' in x):
        return 1
    else:
        return 0
ramen_df4['beef'] = ramen_df4['beef'].apply(beef_replace)


#big or hearty
ramen_df4['big']= ramen_df4['Variety'].copy()
def big_replace(x):
    if ('Big' in x or 'Hearty' in x):
        return 1
    else:
        return 0
ramen_df4['big'] = ramen_df4['big'].apply(big_replace)

#cream
ramen_df4['cream']= ramen_df4['Variety'].copy()
def cream_replace(x):
    if ('Cream' in x or 'Creamy' in x or 'Carbonara' in x):
        return 1
    else:
        return 0
ramen_df4['cream'] = ramen_df4['cream'].apply(cream_replace)

#diff_noodle
ramen_df4['diff_noodle']= ramen_df4['Variety'].copy()
def noodle_replace(x):
    if ('Udon' in x or 'Soba' in x or 'Vermicelli ' in x):
        return 1
    else:
        return 0
ramen_df4['diff_noodle'] = ramen_df4['diff_noodle'].apply(noodle_replace)


#color?

#to create spiciness pivot table
#ramen_df4['Spiciness']= ramen_df4['spicy'].copy()
#def sp_replace(x):
#    if x == 1:
#        return 'spicy'
#    else:
#        return 'non-spicy'
#ramen_df4['Spiciness'] = ramen_df4['Spiciness'].apply(sp_replace)


ramen_df4.head(1000)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


x = ramen_df4[['Cup','Pack','Tray','Bowl','Australia','Canada','China','Germany','Hong Kong','India','Indonesia','Japan','Malaysia','Philippines', 'Singapore', 'South Korea','USA','spicy', 'seafood','premium','miso','Nissin','Koka','Nongshim','Waiwai','Yamamoto','Yamamori','Yamadai','fried','artificial','Indomie','vegetarian','pork','chicken','beef','big','cream']]

y = ramen_df4[['Stars']]

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)

lm = LinearRegression()

lm.fit(x_train, y_train)
y_predict = lm.predict(x_test)

# Input code here:

print("Train score:")
print(lm.score(x_train,y_train))

print("Test score:")
print(lm.score(x_test,y_test))

residuals = y_predict - y_test
plt.scatter(y_predict, residuals, alpha=0.4)
plt.title('Residual Analysis')

plt.show()
```

```python
predictions = lm.predict(x_test)
sns.distplot((y_test-predictions),bins=50);
```

```python

model = lm.fit(x_train, y_train)

y_predict= lm.predict(x_test)

print("Train score:")
print(lm.score(x_train, y_train))

print("Test score:")
print(lm.score(x_test, y_test))

plt.scatter(y_test, y_predict, alpha=0.1)
plt.plot(range(5), range(5))

plt.xlabel("Stars: $Y_i$")
plt.ylabel("Predicted Stars: $\hat{Y}_i$")
plt.title("Stars vs Predicted Stars using MLR")

plt.show()
```

```python
#lm.coef_
#['Cup','Pack','Tray','Bowl', 'Australia','Canada','China','Germany','Hong Kong','India','Indonesia','Japan','Malaysia']
list(zip(model.coef_, x))
```

```python
from sklearn.neighbors import KNeighborsRegressor

kn_regressor = KNeighborsRegressor(n_neighbors = 50, weights = "distance")
kn_regressor.fit(x_train, y_train)

knr_training_score = kn_regressor.score(x_train, y_train)

print("KNR training set score: %s" %(knr_training_score))

knr_test_score = kn_regressor.score(x_test, y_test)

print("KNR test set score: %s" %(knr_test_score))

#generate plot showing difference between predicted age and actual age using K Neighbors Regressor

knr_predictions = kn_regressor.predict(x_test)

plt.scatter(y_test, knr_predictions,  alpha=0.1)
plt.title("Actual star versus predicted star using K Neighbors Regression")
plt.xlabel("Actual star")
plt.ylabel("Predicted star")
plt.show()
```
