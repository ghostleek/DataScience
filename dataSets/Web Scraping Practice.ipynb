```python
# get date references
from datetime import timedelta, date

dates =[]
dates_2016=[]
dates_2017=[]
dates_2018=[]
dates_2019=[]
def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_date_2016= date(2016,1,1)
end_date_2016= date(2017,1,1)
start_date_2017= date(2017,1,1)
end_date_2017= date(2018,1,1)
start_date_2018= date(2018,1,1)
end_date_2018= date(2019,1,1)
start_date_2019= date(2019,1,1)
end_date_2019= date(2020,1,1)


for single_date in daterange(start_date_2016, end_date_2016):
  dates_2016.append(single_date.strftime("%Y-%m-%d"))

for single_date in daterange(start_date_2017, end_date_2017):
  dates_2017.append(single_date.strftime("%Y-%m-%d"))

for single_date in daterange(start_date_2018, end_date_2018):
  dates_2018.append(single_date.strftime("%Y-%m-%d"))

for single_date in daterange(start_date_2019, end_date_2019):
  dates_2019.append(single_date.strftime("%Y-%m-%d"))

dates_2019
```

```python
results = []
timestamp = []
base_url= 'https://api.data.gov.sg/v1/environment/rainfall?date='
for date in dates_2016:
  res = requests.get(base_url+date)
  for i in range(len(res.json()['items'])):
    #print(res.json()['items'][i]['timestamp'])
    time = res.json()['items'][i]['timestamp']
    timestamp.append(time)
    # we'll take only the pm25 readings
    #print(res.json()['items'][i]['readings']['pm10_sub_index'])
    results.append(pd.DataFrame(res.json()['items'][i]['readings']['pm25_sub_index'].items()).set_index(0).T)
  
psi_df_2016 = pd.concat(results)

psi_df_2016.reset_index(drop = True, inplace = True)
psi_df_2016['timestamp'] = timestamp
psi_df_2016.head(100)
#df_concat.reset_index(drop = True, inplace = True)
#df_concat['timestamp'] = timestamp
#df_concat.head()
```
